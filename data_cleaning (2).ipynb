# Detecting the encoding type of my csv file by using chardet library

## (chardet is a Universal encoding detector for Python 3)


```python
import chardet

# Detect the encoding of the CSV file
with open('globalterrorism.csv', 'rb') as rawdata:
    result = chardet.detect(rawdata.read())
encoding = result['encoding']

print(f"The detected encoding is: {encoding}")
```

```python
# Calculate missing values
missing = df.isnull().sum()

# Iterate through the missing values and print them
for column, count in missing.items():
    print(f'Column: {column}, Missing Values: {count}')
```

```python
threshold = 0.5  # For example, to keep columns with less than 50% missing values

# Calculate the number of missing values in each column
missing_values = df.isnull().sum()

# Calculate the percentage of missing values in each column
percentage_missing = (missing_values / len(df)) * 100

# Create a list of columns to drop based on the threshold
columns_to_drop = percentage_missing[percentage_missing > threshold].index

# Drop the selected columns from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
missing = df.isnull().sum()

# Iterate through the missing values and print them
for column, count in missing.items():
    print(f'Column: {column}, Missing Values: {count}')
```
